**Import the vizualization library (1 mark)**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**Display cor-relation matrix (1 mark)**

```python
df_numeric_cancer=df.select_dtypes(exclude=object)
df_numeric_cancer.corr()
```

```python
sns.set(rc={'figure.figsize':(17,6.5)})
sns.heatmap(df_numeric_cancer.corr(),annot=True)
```

**Display the histogram of the dataframe (2 mark)**

**Display the countplot for Gender at different Level (1 mark)**

```python
# count plot on single categorical variable
sns.countplot(x ='Gender', hue='Level', data = df)
 
# Show the plot
plt.show()
#Number of Gender ‘1’ is observed to be high
```

**Display the countplot for Gender at different Smoking (1 mark)**

```python
# count plot on single categorical variable
sns.countplot(x ='Gender', hue='Smoking', data = df)
 
# Show the plot
plt.show()
#Number of smokers who has cancer are more, especially in male
```

**Display a pie chart the % of different Level (1 mark)**

```python
level_counts=df["Level"].value_counts().sort_index()
level_percentage=level_counts/level_counts.sum()
plt.pie(level_percentage,labels=level_percentage.index,autopct='%.0f%%')
plt.show()
#Cancer patients at different levels are almost equally distributed
```
